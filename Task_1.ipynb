```python
import pandas as pd

df = pd.read_csv('data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
```

```python
sales_category = df.groupby('Category')['Sales'].sum()

sales_category.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
profit_category = df.groupby('Category')['Profit'].sum()

profit_category.plot(kind='bar')

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title('Sales by Region')
plt.ylabel('')
plt.show()
```

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)

top_states.plot(kind='bar')

plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Sales')
plt.show()
```

```python
sns.scatterplot(data=df, x='Discount', y='Profit')

plt.title('Discount vs Profit')
plt.show()
```

The Superstore dataset was cleaned and analyzed using Python. Sales and profit trends were identified across categories, regions, and states. The analysis revealed that Technology products generate the highest revenue and profit, while excessive discounts negatively impact profitability. Data visualization helped uncover valuable business insights that can support better decision-making.
